## 3. Funciones

Se definen con la palabra reservada def, el nombre de la función y unos paréntesis, que también se utilizan para hacer la llamada y para pasar los argumentos si la función los requiere.

```python
def saludar():
    print("Hola! Este print se llama desde la función saludar()")

saludar()  #Llamamos a la funcion
```

Dentro de una función podemos utilizar variables y sentencias de control.

```python
def repetir():
    for i in range(3):
        print("Frase repetida ",[i])
        
repetir()
```

Ahora le pasamos el numero de veces que queremos que se repita como argumento.

```python
def repetir(n_veces):
    for i in range(n_veces):
        print("Frase repetida ",[i])

n_veces = 5
repetir(n_veces)
```

### Ambito de las variables

Algo muy importante a tener en cuenta cuando tratamos con funciones es el ambito de las variables.

Una variable declarada fuera de la función (al mismo nivel), sí que es accesible desde la función.

```python
variable_fuera = 'Soy una variable fuera de la funcion'

def funcion():
    print(variable_fuera)

funcion()
```

```python
def funcion():
    variable_dentro = 'Soy una variable dentro de la funcion'
    print(variable_dentro)

funcion()
```

En cambio una variable declarada dentro de la funcion, no puede ser llamada fuera de ella.

```python
def funcion():
    variable_dentro = 'Soy una variable dentro de la funcion'
    print(variable_dentro)

print(variable_dentro)
funcion()
```

### Solución
Primero agrupemos los datos por sitio y por sexo y después calculemos el total para cada sitio.

```python
by_site_sex = surveys_df.groupby(['plot_id','sex'])
site_sex_count = by_site_sex['weight'].sum()
site_sex_count
```

Ahora usaremos .unstack() en los datos agrupados para entender como el peso total de cada sexo contribuye a cada sitio.

```python
by_site_sex = surveys_df.groupby(['plot_id','sex'])
site_sex_count = by_site_sex['weight'].sum()
site_sex_count.unstack()
```

Ahora creamos una gráfica de barras apilada con los datos donde el peso para cada sexo es apilado por sitio.


En vez de mostrarla como tabla, nosotros podemos graficar los datos apilando los datos de cada sexo como sigue.

```python
spc = site_sex_count.unstack()
s_plot = spc.plot(kind='bar',stacked=True,title="Total weight by site and sex")
s_plot.set_ylabel("Weight")
s_plot.set_xlabel("Plot")
```

```python
# Ejemplo de la pagina https://unipython.com/matplotlib-funciones-principales/

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
fig = plt.figure()
# Tipo de figura
ax = fig.gca(projection='3d')
# Datos
X = np.arange(-4, 4, 0.3)
Y = np.arange(-4, 4, 0.3)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
# Graficamos o trazamos la superficie
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='jet')
# Personalizamos el ejex z
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
# Agregamos una barra de colores para que asigne valores a los colores.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
