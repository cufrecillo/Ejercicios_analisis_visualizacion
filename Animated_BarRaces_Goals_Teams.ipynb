```python
#funcion que devuelve un dataframe con los goles del equipo tanto si es local como visitante

def dataFrameTeamBBDD(dataframe, team):
    df_team_1 = pd.DataFrame()
    df_team_2 = pd.DataFrame()
    
    for i in range (0, len(dataframe)):
        if dataframe["HomeTeam"][i] == team:
            df_aux = dataframe.iloc[[i], dataframe.columns.get_indexer(["Date", "HomeTeam", "FTHG"])]
            df_team_1 = df_team_1.append(df_aux,ignore_index=False)
        
        elif dataframe["AwayTeam"][i] == team:
            df_aux2 = dataframe.iloc[[i], dataframe.columns.get_indexer(["Date", "AwayTeam", "FTAG"])]
            df_team_2 = df_team_2.append(df_aux2,ignore_index=False)
        
    df_team_1.rename(columns= {'HomeTeam': 'Team', 
                                  'FTHG': 'Goals'}, inplace=True)

    df_team_2.rename(columns= {'AwayTeam': 'Team', 
                                    'FTAG': 'Goals'}, inplace=True)

    dataframe = pd.concat([df_team_1, df_team_2])
    #dataframe.reset_index(drop=True, inplace=True)
    
    return dataframe
```

```python
# lista para seleccionar los equipos que se van analizar sus goles
list_teams = ['Liverpool', 'Man City', 'Barcelona', 'Real Madrid', 
              'Bayern Munich', 'Juventus', 'Paris SG']
```

```python
df_team1 = dataFrameTeamBBDD(df, list_teams[0])
df_team2 = dataFrameTeamBBDD(df, list_teams[1])
df_team3 = dataFrameTeamBBDD(df, list_teams[2])
df_team4 = dataFrameTeamBBDD(df, list_teams[3])
df_team5 = dataFrameTeamBBDD(df, list_teams[4])
df_team6 = dataFrameTeamBBDD(df, list_teams[5])
df_team7 = dataFrameTeamBBDD(df, list_teams[6])
```
